```python
# Calculate the sum of squared residuals
SSres = np.sum((final_commodity["Export_Value"] - final_commodity["Import_Value"])**2)

# Calculate the total sum of squares
SStot = np.sum((final_commodity["Export_Value"] - final_commodity["Export_Value"].mean())**2)

# Calculate R-squared
R_squared = 1 - (SSres / SStot)

# Calculate adjusted R-squared
adjusted_R_squared = 1 - (1 - R_squared) * (final_commodity.shape[0] - 1) / (final_commodity.shape[0] - 2 - 1)

# Print the results
print("R-squared: ", R_squared)
print("Adjusted R-squared: ", adjusted_R_squared)
```

```python
# Split the data into the independent variable (export value) and the dependent variable (trade deficit)
X = final_commodity.drop(["year","Commodity","Trade Deficit"],axis = 1)
y = final_commodity['Trade Deficit']
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
# Define the models and their hyperparameter grid
model = {
    'linear_regression': {'model': LinearRegression(), 'params': {'positive': [True, False]}},
    'lasso_regressor': {'model': Lasso(), 'params': {"alpha": [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}},
    'ridge_regressor': {'model': Ridge(), 'params': {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}},
    'random_forest': {'model': RandomForestRegressor(), 'params': {'n_estimators': [50, 100, 200, 300]}},
    'decision_tree': {'model': DecisionTreeRegressor(), 'params': {'max_depth': [3, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50], 'splitter': ['best', 'random']}},
    'elastic_net': {'model': ElasticNet(), 'params': {"alpha": [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}},
    "ada_boosting": {"model": AdaBoostRegressor(), 'params': {'n_estimators': [50, 100, 200, 300]}},
    "support_vector": {"model": SVR(), 'params': {"kernel": ['linear', 'poly', 'rbf', 'sigmoid']}}
}
def find_best_model_using_gridsearchcv(X, y):
    scores = []
    cv = ShuffleSplit(n_splits=2, test_size=0.2, random_state=0)
    for model_name, config in model.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': model_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })


    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

final_model = find_best_model_using_gridsearchcv(X,y)
print(final_model)

# Make predictions for all the models
predictions = {}
for model_name, config in model.items():
    gs = GridSearchCV(config['model'], config['params'], cv=2, return_train_score=False)
    gs.fit(X_train, y_train)
    predictions[model_name] = gs.predict(X_test)

# Print the predictions
print(predictions)

# Pickle the models
for model_name, predictions in predictions.items():
    with open(f"{model_name}.pkl", "wb") as f:
        pickle.dump(predictions, f)
```
